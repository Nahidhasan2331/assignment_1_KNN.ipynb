```python
def normalize_images(images):
    images = images.astype(np.float32) / 255.0
    return images
```

```python
knn_model = KNeighborsClassifier(n_neighbors=5)  # Adjust n_neighbors as desired

def l1_distance(a, b):
    return np.sum(np.abs(a - b))

def l2_distance(a, b):
    return np.sqrt(np.sum(np.square(a - b)))
```

```python
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier

class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']
labels = []

num_samples_per_class = 5000

# Populate labels with class labels
for class_name in class_names:
    labels.extend([class_name] * num_samples_per_class)  # Replace `num_samples_per_class` with the actual number of samples per class

labels = np.array(labels)

# Create a NumPy array of random numbers
images = np.random.rand(50000, 784)

k_values = [1, 3, 5, 7, 9]  # Adjust as desired

for k in k_values:
    knn_model = KNeighborsClassifier(n_neighbors=k)
    knn_model.fit(images, labels)

    l1_scores = cross_val_score(knn_model, images, labels, cv=5, scoring="accuracy", n_jobs=-1, verbose=1, error_score="raise")
    l2_scores = cross_val_score(knn_model, images, labels, cv=5, scoring="accuracy", n_jobs=-1, verbose=1, error_score="raise")

    average_l1_score = np.mean(l1_scores)
    average_l2_score = np.mean(l2_scores)

    print(f"For k={k}, Average L1 Score: {average_l1_score:.4f}, Average L2 Score: {average_l2_score:.4f}")

import matplotlib.pyplot as plt

plt.plot(k_values, average_l1_scores, label="Manhattan (L1)")
plt.plot(k_values, average_l2_scores, label="Euclidean (L2)")
plt.xlabel("K")
plt.ylabel("Accuracy")
plt.title("Accuracy for Different K values and Distance Metrics")
plt.legend()
plt.show()

```

```python
import glob

test_image_paths = glob.glob(r"D:\Nahid\CIFAR-10-images-master\CIFAR-10-images-master\test\airplane\*.jpg")

# Load test images and preprocess them
test_images = []
for image_path in test_image_paths:
    if os.path.isfile(image_path):
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        test_images.append(image)

# Create a new variable to store the first 1000 images
test_images_small = test_images[:1000]

# Convert the list of images to a NumPy array
test_images_small = np.array(test_images_small)

# Preprocess the test images
test_images_small = normalize_images(test_images_small)

# Flatten the test images
test_images_flat_small = test_images_small.reshape(test_images_small.shape[0], -1)

# Create a copy of the flattened test images
test_images_flat_small_copy = np.zeros_like(test_images_flat_small)

# Initialize the copy with the data from the original array
test_images_flat_small_copy[:] = test_images_flat_small

# Reshape the test images to have dimensions (num_samples, num_features)
test_images_flat_small_copy = test_images_flat_small_copy.reshape(test_images_flat_small_copy.shape[0], 784)

# Predict the labels of the test images
predictions = knn_model.predict(test_images_flat_small_copy)

# Display predictions
for i, image_path in enumerate(test_image_paths[:1000]):
    print(f"Test Image: {image_path}")
    print("Top 5 Predictions:")

    prediction_indices = np.argsort(-knn_model.predict_proba(test_images_flat_small_copy[i].reshape(1, -1))[0])[:5]
    for index in prediction_indices:
        class_name = class_names[index]
        confidence = knn_model.predict_proba(test_images_flat_small_copy[i].reshape(1, -1))[0][index]
        print(f"Class: {class_name}, Confidence: {confidence:.4f}")
```
